```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def gaussian_kernel(size, sigma):
    if size % 2 == 0:
        size = size + 1
    max_point = size // 2

    min_point = -max_point

    K = np.zeros((size, size))  # kernel matrix
    for x in range(min_point, max_point + 1):
        for y in range(min_point, max_point + 1):
            value = (
                1
                / (2 * math.pi * sigma * sigma)
                * math.exp((-1) * ((x * x + y * y) / (2 * sigma * sigma)))
            )
            K[x - min_point, y - min_point] = value

    return K


img = cv2.imread("../sources/images/2.jpg", 0)

img_gaussian_25 = cv2.filter2D(img, ddepth=-1, kernel=gaussian_kernel(5, 2.5))
img_gaussian_50 = cv2.filter2D(img, -1, gaussian_kernel(5, 5.0))
img_gaussian_100 = cv2.filter2D(img, -1, gaussian_kernel(5, 10.0))
```
