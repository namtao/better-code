```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đọc hiển thị ảnh

img1 = cv2.imread("../sources/images/1.jpg")

plt.imshow(img1, cmap = 'gray')
```

```python
# Chuyển đổi không gian màu

img_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap = 'gray')
```

```python
# Tăng độ sáng 50 đơn vị

img_ = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

new_img = img_.astype(np.float32) + 50
new_img = np.clip(new_img, 0, 255)
new_img = new_img.astype(np.uint8)

plt.imshow(new_img)
```

```python
# Giảm độ sáng 80 đơn vị

img_ = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

new_img = img_.astype(np.float32) - 80
new_img = np.clip(new_img, 0, 255)
new_img = new_img.astype(np.uint8)

plt.imshow(new_img)
```
