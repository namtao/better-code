```python
text ="Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối cùng tháng 2, có bản miễn phí với 10 bài hát mỗi ngày"
```

```python
# Gán nhãn từ loại

from underthesea import pos_tag

pos_tag(text)
```

```python
# Phân tích văn bản

from underthesea import classify

classify(text)
```

```python
# Phân tích cảm xúc

from underthesea import sentiment
sentiment(text)
```

```python
# Phân đoạn câu văn

from underthesea import sent_tokenize

sent_tokenize(text)
```

```python
# Phân đoạn từ ngữ

from underthesea import word_tokenize

word_tokenize(text)
```
